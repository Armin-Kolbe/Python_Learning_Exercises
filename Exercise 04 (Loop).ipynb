```python
txt = input("Please enter your text: ")    # for example: I can %not eat 23 @bread# and 34 balls 
amount1 = amount2 = 0
for i in range(int(len(txt))):
  if txt[i:i+1].isalpha():
    amount1 += 1
  elif txt[i:i+1].isdigit():
    amount2 += 1
print(f"the txt hat {amount1} letters and {amount2} digits and {len(txt)-amount1-amount2} invalid characters")

```

    <h4>Question 03</h4>
<p>Write a program to analyze the scores of 10 students in a class. You need to
collect the name and score (out of 20) for each student. Based on their
scores, you need to assign them a level as follows: 
</p>

<ul>
    <li>A+ : If the score is between 17 and 20 (inclusive)</li>
    <li>A : If the score is between 15 and less than 17</li>
    <li>B : If the score is between 10 and less than 15</li>
    <li>C : If the score is less than 10</li>
</ul>

<p>
    Your program should display the following:
</p>

<ul>
    <li>The name, score, and level of each student.</li>
    <li>The name and score of the student with the highest score.</li>
    <li>The name and score of the student with the lowest score.</li>
    <li>The average score of the class.</li>
    <li>
        Make sure to validate the input to ensure that the score is a positive number and less than
        or equal to 20.
    </li>
</ul>

</div>

```python
name = []
score = []
level = []
for j in range(10):  
  name.append("")
  score.append("")
  level.append("")
for i in range(10):
  name[i] = input(f"Please enter the name of Student {i+1}: ")      
  score[i] = input(f"Please enter the score of Student {i+1}: ")  
  if float(score[i]) > 20 or float(score[i]) < 0:
    print("error: The score must be between 0 and 20")
    break  
  elif float(score[i]) > 17 :
    level[i] = "A++"
  elif float(score[i]) > 15 :
    level[i] = "A"
  elif float(score[i]) >= 10 :
    level[i] = "B"
  else:
    level[i] = "C"
print("---------------------------------------------------")
print( "The list of all students:")
for k in range(10): 
  print(f"{name[k]} => score = {score[k]} , level = {level[k]}")
print(" ")
print("---------------------------------------------------")
print( "The list of the students with the highes score:")
amount = 0
for l in range(10):
  if level[l] == "A++":
    print(name[l], end="    ")
    amount +=1
if amount == 0:
  print(" ")
  print(">>> There is no students with the highes score")
print(" ")
print("---------------------------------------------------")
print( "The list of the students with the lowest score:")
amount = 0
for t in range(10):
  if level[t] == "C":
    print(name[t], end="    ")
    amount +=1
if amount == 0:
  print(" ")
  print(">>> There is no students with the lowest score")
print(" ")
print("---------------------------------------------------")
sum_students = 0
sum_scores = 0
for p in range(10):
  sum_scores += float(score[p])
  sum_students += 1
if sum_students != 0 :
  av = round(sum_scores/sum_students, 2)
  print ("The Average score of all students is: ", av)
else: 
  print ("you entered no students with score. Average Can not be count!")


```
